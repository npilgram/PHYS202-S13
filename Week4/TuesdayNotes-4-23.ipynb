Electirc Dipole Potential
--

$V(x,y)=\frac{kq}{\sqrt{x^2+(y-d)^2}}-\frac{kq}{\sqrt{x^2 + (y+d)^2}}$

```python
#make the x-y grid to cover the plane -2<x<x and -2<y<2
dx = 0.06
dy = 0.06
x,y = meshgrid(arange(-2.,2.,dx),arange(-2.,2.,dy))
print x.ndim
print y.ndim
print x.shape
print y.shape
#creates grid array where one row stays constant and the others varry
```

```python
k=1
q=1
d=1
Vxy = (k*q)/sqrt((x**2)+(y-d)**2)-(k*q)/sqrt((x**2)+(y+d)**2)
print Vxy
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = figure()
ax = fig.gca(projection='3d') #gets axis
p = ax.plot_surface(x, y, Vxy, rstride=1, cstride=1, linewidth=0, cmap=cm.flag)
cb = fig.colorbar(p, shrink=0.5)
```

```python
help(Axes3D.plot_surface)
```

```python
levels = array([8, 4, 2, 0.5, 0.25, 0, -0.25, -0.5, -1, -2, -4, -8])
contour(x, y, Vxy, levels, cmap=cm.flag, linewidths=5)
show()
```

```python
def pointPotential(x,y,q,posx,posy):
    """Returns the electric potential at point (x,y) created by the charge of q at (posx,posy). Expects units of meters for x, y, posx, and posy, and units of Coulombs for q."""
    k = 8.9875517873681764*10.**9. #Nm^2/C^2
    Vxy = (k*q)/(sqrt(((x-posx)**2)+((y-posy)**2)))
    return Vxy

    
```

```python
x=1.; y=1.
V11= pointPotential(x,y,1e-9,0.,0.)
print V11
```

```python
x,y = meshgrid(arange(-2.,2.,0.06), arange(-2.,2.,0.06))
Vxy = pointPotential(x,y,1e-9,0.,0.)
print Vxy
```

###b.)

```python
def dipolePotential(x,y,q,d):
    """Returns the electrical potential at point (x,y) created by the dipole with chage magnitudes of q and charge separation of d. Expects units of meters for x, y, and d and units of Coulombs for q"""
    Vxy = pointPotential(x,y,q,-d/2.,0.)-pointPotential(x,y,q,d/2.,0.)
    return Vxy
    
```

###c.)

## Dipole along x-axis

```python
x,y = meshgrid(arange(-2.,2.,0.06),arange(-2.,2.,0.06))
q=1e-9 #C
d=2 #m
Vxy = dipolePotential(x,y,q,d)
fig = figure()
ax = fig.gca(projection='3d')
v = ax.plot_surface(x,y,Vxy, rstride=1, cstride=1, linewidth=0, cmap=cm.coolwarm)
cb = fig.colorbar(v, shrink=0.5)
```

## Quadrapole

```python
dx = 8./101
dy = 8./101
x,y= meshgrid(arange(-4,4,dx), arange(-4,4,dy))
q_1 = 1e-9
q_2 = -1e-9
Vxy = pointPotential(x,y,q_1,0.,2.)+pointPotential(x,y,q_1,0.,-2.)+pointPotential(x,y,q_2,2.,0.)+pointPotential(x,y,q_2,-2.,0.)
fig = figure()
ax = fig.gca(projection='3d')
v = ax.plot_surface(x,y,Vxy, rstride=1, cstride=1,cmap=cm.coolwarm, linewidth=0)
cb = fig.colorbar(v, shrink=0.5)
```
