creat a very simple plot

```python
t= arange(0.0, 1.0+0.01, 0.01)
c = cos(2*2*pi*t)
s = sin(2*2*pi*t)
```

```python
print pi
```

```python
plot(t, c, color='blue', linestyle=':', linewidth=4, label="cos(4${\pi}$t)")
plot(t, s, color='red', linewidth=3, label="sin(4${\pi}$t)")
xlabel("time (s)", fontsize=20)
ylabel("voltage (mV)", fontsize=20)
#add title for convenience
title('A simple plot', fontsize=30)
grid(True)
legend(loc="lower left")
#savefig('simple_plot.pdf')
```
