##Approximate Soultion

```python
g=9.8 #m/s^2
l=0.5 #m
th0 = 20.*pi/180 # 20 degrees but must use radians

t = arange(0., 10.+0.01, 0.01)
theta_app = lambda t: th0*cos(sqrt(g/l)*t)

plot(t, theta_app(t))
show()
```

##Numerical Solution

```python
def deriv(theta_current,t): #returns derivatve of the array theta
    g=9.8 #m/s^2
    l=0.5 #m
    theta = theta_current[0]
    omega = theta_current[1]
    dtheta = omega
    domega = -(g/l)*sin(theta)
    return array([ dtheta, domega ])
    
```

```python
from scipy.integrate import odeint
#odeint?
```

```python
t = arange(0., 10.+0.01, 0.01)
omega0 = 0. #starting from rest, omega=0
theta_int = array([th0, omega0])

theta= odeint(deriv, theta_int,t)
```

```python
theta.shape
```

```python
plot(t, theta_app(t), 'b-', t, theta[:,0], 'r--')
show()
```

```python
th0 = 50.*pi/180 #50 degrees

theta_int = array([th0, omega0])
theta= odeint(deriv, theta_int,t)

plot(t, theta_app(t), 'b-', t, theta[:,0], 'r--')
show()
```

```python
th0, th1, th2, th3, th4 = 0, (pi/4), (pi/2), ((3*pi)/4), pi

theta_int0 = array([th0, omega0])
theta0= odeint(deriv, theta_int0,t)

theta_int1 = array([th1, omega0])
theta1= odeint(deriv, theta_int1,t)

theta_int2 = array([th2, omega0])
theta2= odeint(deriv, theta_int2,t)

theta_int3 = array([th3, omega0])
theta3= odeint(deriv, theta_int3,t)

theta_int4 = array([th4, omega0])
theta4= odeint(deriv, theta_int4,t)

plot(t, theta0[:,0],'b-', t, theta1[:,0], 'r--', t, theta2[:,0], 'g:', t, theta3[:,0],'k-.', t, theta4[:,0])
show()
```

period from pi/4 to 3pi/4 gets smaller and the pedulum does not swing with initial angle of 0 or pi
