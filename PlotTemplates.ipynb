Publication ready plot template
==
Simple graphing template for matplotlib functions and data plots

Line Plots
--

```python
y=arange(-5,5,0.01) #this creates the range of values over wich functions of y will be plotted
E=cosh(y) #sets the function of y to be plotted
p= sinh(y)
plot(y,E, color='blue', linestyle='-', linewidth=2, label='$sinh(y)$') #plots cosh(y) as a blue line of width 2 and labels it for legend
plot(y,p, color='red', linestyle='--', linewidth=3, label='$sinh(y)$') #plots sinh(y) as red dashed line of width 3 and labels it
xlabel('$x-axis$', fontsize=20) # labels the x axis with font size of 20, $'s make it italicized
ylabel('$y-axis$', fontsize=20) # labels the y axis with font size of 20, $,s make it italicized
grid(True) # adds gid lines
legend(loc='lower right') # adds legend in lower right
title('Plot', fontsize=30) # adds title, font size of 30
```

Scatter Plots
---

```python
n=200
X = random.normal(0,1,n)
Y = random.normal(0,1,n) #random number data for example plot
```

```python
plot(X, Y,'r.', label='Scatter Plot') #creats scatter plot, the 'r.' plots the points
xlabel('x-axis', fontsize=20)
ylabel('y-axis', fontsize=20) # labels x and y axis
grid(True) #adds grid
legend(loc='lower left') #adds legend in lower left
title('Scatter Plot', fontsize=30) #adds title
# xlim() and ylim() can be used to flip data by entering the bound in reverse
```
