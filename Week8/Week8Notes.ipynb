Sampling from a Continuous Distribution
--

Cumulative Distriburion Function (CDF)
-

Example 1
--

```python
x = arange(0,pi+0.001,0.01)
f = lambda x: sin(x)/2
F = lambda x: (cos(0)-cos(x))/2
    
plot(x,f(x), label='True distribution', lw=3)
plot(x,F(x), label='Cumulative function', lw=3)
     
xlim(0,3.5)
ylim(0,1.1)

legend(loc=0)
show()
```

```python
#Number of samples
N=10000

#random numbers between 0 and 1
z = numpy.random.random_sample(N);

#Inverse CDF
Finv = lambda z: arccos(cos(0)-2*z)
    
#sampled x values
xsamples = Finv(z)

#how do they compare to the original
hist(xsamples, 50, normed=True, label='Sampled distribution')
plot(x, f(x), label='True distribution', lw=3, color='red')

xlim(0,pi)
ylim(0,0.7)

legend(loc=0)
show()
```

###a.)

$f(x)=\lambda e^{-\lambda x}$

$F(x)=1-e^{-\lambda x}$

$F^{-1}(x)=-\frac{1}{\lambda}ln(1-x)$

###b.)

```python
x = arange(0., 5.+0.001, 0.01)
C=1
f= lambda x: C*e**(-C*x)
F= lambda x: 1-e**(-C*x)
    
plot(x, f(x), lw=3, label='True distribution')
plot(x, F(x), lw=3, label='CDF')
legend(loc=0)
show()
```

###c.)

```python
N=10000

z=numpy.random.random_sample(N);

Finv= lambda z: (-1/C)*log(1-z)

xsamples=Finv(z);

hist(xsamples,50,normed=True, label='Sampled distribution')
plot(x,f(x),lw=3,color='red', label='True distribution')
xlim(0,5)
legend(loc=0)
show()
```
